```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

```python
# Definindo a função apresentada no enunciado
def N(m, N0, mc):
    return N0 * np.exp(0.307*m)*(1 - np.exp(3*(mc - m)))

# Colocando os valores da tabela
m_valores = np.array([20.50, 20.60, 20.75, 20.90, 21.10, 21.25, 21.40, 21.50, 21.70])
N_valores = np.array([3, 5, 12, 20, 18, 35, 25, 30, 20])
```

```python
# Fazendo o ajuste da curva
p0 = [21,22]
popt, pcov = curve_fit(N, m_valores, N_valores, p0=p0)
```

```python
Nopt, mcopt = popt
print('O valor optimizado para N0 é: ', Nopt)
print('O valor optimizado para mc é: ', mcopt)
print('E os erros desses valores são: ')
print(pcov)
```

```python
# a magnitude de corte da função de luminosidade das nebulosas planetárias é:
Mc = -4.47

# temos que: m − M = 5 log(d) − 5
d = 10**((mcopt - Mc + 5)/5)
d = d / 1e6
print('Distância em Mpc: ', round(d, 2))
```

```python
# Plotando os dados:
plt.figure()
plt.plot(m_valores, N_valores, 'o', color='darkred', label='Dados da tabela')
plt.plot(m_valores, N(m_valores, Nopt, mcopt), color='darkblue', label='Ajuste calculado')
plt.xlabel('m')
plt.ylabel('N')
plt.grid(True)
plt.legend(loc='upper left')
plt.show()
```
